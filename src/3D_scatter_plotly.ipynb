```python
# -*- coding: utf-8 -*-
"""
Created on Tue Aug 20 12:31:39 2024

@author: marko
"""

import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns
import os
from os.path import join
import pandas as pd


current_path = os.getcwd()

#path to pickle file
pk_file = "C:/Users/marko/Downloads/files_5k/data_clustered_5kentries.pkl"

pk_path = join(current_path, pk_file)

# Step 1: Load the data
with open(pk_path, 'rb') as f:
    data = pickle.load(f)

# Convert the list of histograms into a proper 2D NumPy array
histograms = np.array(data['RGB_Histogram'])  # Initially, this might be an object array
histograms = np.vstack(histograms)  # Convert it into a 2D array of shape (5929, 768)

# Extract other relevant data
image_ids = np.array(data['Image_ID'])        # Corresponding image IDs
clusters = np.array(data['cluster_histogram'])# Corresponding cluster labels

# Ensure the shapes are as expected
print(f"Histograms shape after reshaping: {histograms.shape}")  # Should now be (5929, 768)
print(f"Image IDs length: {len(image_ids)}")    # Should match 5929
print(f"Clusters length: {len(clusters)}")      # Should match 5929

# Count the number of unique clusters
unique_clusters = np.unique(clusters)
num_clusters = len(unique_clusters)
print(f"Number of clusters: {num_clusters}")

# Step 2: Apply UMAP or TSNE for dimensionality reduction
# We'll use UMAP first

# UMAP
#reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
#embedding = reducer.fit_transform(histograms)

# Optionally, you can try t-SNE instead
tsne = TSNE(n_components=3, random_state=0,early_exaggeration=50)
embedding = tsne.fit_transform(histograms)

# Convert clusters to categorical type
clusters = pd.Categorical(clusters, categories=sorted(unique_clusters))

print("dimensionality reduction finished")
```
